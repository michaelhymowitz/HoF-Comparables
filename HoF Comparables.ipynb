```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# ppl is a dataframe for every MLB players' names
ppl_ = pd.read_csv("People.csv") # ppl_ is created to make later work easier
keep_cols = ["playerID", "nameGiven", "nameLast", ]
ppl_ = ppl_.loc[:, keep_cols]
ppl_["name"] = ppl_["nameGiven"] + " " + ppl_["nameLast"]

ppl = ppl_.set_index("playerID")
ppl = ppl.loc[:, ["name"]]
ppl.head()
```

```python
# Making a dictionary and list of all HoFamers
id_dict = ppl.to_dict()
id_dict = id_dict["name"]
id_list = id_dict.keys()
id_dict
```

```python
# fd_dict will let us create separate dataframes for position players and pitchers
pos = pd.read_csv("Fielding.csv")
pos = pos[pos["yearID"] >= 1903].reset_index(drop=True)
pos = pos.loc[:, ["playerID", "POS"]]
pos = pos.set_index("playerID")
pos_dict = pos.to_dict()
pos_dict = pos_dict["POS"]
pos_dict
```

```python
# Dataframes for position players and pitchers
pos_htrs = pos[pos["POS"] != "P"]
pos_ptrs = pos[pos["POS"] == "P"]
```

```python
# bt is a dataframe for all hitting seasons of MLB players, 1961 marks beginning of Expansion Era
pd.set_option("display.max_columns", None)
bt = pd.read_csv("Batting.csv")
bt = bt[bt["yearID"] >= 1961].reset_index(drop=True)
bt.rename(columns = {"SO" : "K"}, inplace = True)
bt.head()
```

```python
bt["playerID2"] = bt["playerID"]
bt.rename(columns = {"playerID": "name", "playerID2" : "playerID"}, inplace = True)
bt = bt.replace({"name": id_dict})
bt["POS"] = ""
bt["POS"] = bt["playerID"]
bt = bt.replace({"POS": pos_dict})
bt.head()
```

```python
# Reordering cols of bt
bt_order = ['name', 'yearID', 'POS', 'teamID', 'lgID', 'G', 'AB', 'R', 'H', '2B',
             '3B', 'HR', 'RBI', 'SB', 'CS', 'BB', 'K', 'IBB', 'HBP', 'SH', 'SF',
             'GIDP', 'playerID']
bt = bt[bt_order]
bt.head()
```

```python
# Cleaning up bt and adding stats
bt["RBI"] = bt["RBI"].astype(int)
bt["SB"] = bt["SB"].astype(int)
bt["CS"] = bt["CS"].astype(int)
bt["K"] = bt["K"].astype(int)
bt["IBB"] = bt["IBB"].astype(int)
bt["HBP"] = bt["HBP"].astype(int)
bt["SH"] = bt["SH"].astype(int)
bt["SF"] = bt["SF"].astype(int)
bt["GIDP"] = bt["GIDP"].astype(int)
bt.head()
```

```python
# Adding stats to bt
bt["1B"] = bt["H"] - (bt["2B"] + bt["3B"] + bt["HR"])
bt["uBB"] = bt["BB"] - bt["IBB"]
bt["BA"] = (bt["H"] / bt["AB"]).round(3)
bt["OBP"] = ((bt["H"] + bt["BB"] + bt["HBP"])
                   / (bt["AB"] + bt["BB"] + bt["HBP"] + bt["SF"])).round(3)
bt["SLG"] = ((bt["1B"] + 2*bt["2B"] + 3*bt["3B"] + 4*bt["HR"]) / bt["AB"]).round(3)
bt["OPS"] = (bt["OBP"] + bt["SLG"]).round(3)
bt["ISO"] = bt["SLG"] - bt["BA"]
bt["BABIP"] = ((bt["H"] - bt["HR"])
                    / (bt["AB"] - bt["K"] - bt["HR"] + bt["SF"])).round(3)
bt["SB%"] = (bt["SB"] / (bt["SB"] + bt["CS"])).round(3)
bt.head()
```

```python
# Reordering the cols of bt to a more intuitive order
bt_col_order = ['name', 'yearID', 'POS', 'teamID', 'lgID', 'G', 'AB', 'R', 'H', '1B', '2B',
                     '3B', 'HR', 'RBI', 'K', 'BB', 'uBB', 'IBB', 'HBP', 'SH', 'SF', 'GIDP', 'SB',
                     'CS', 'SB%', 'BA', 'OBP', 'SLG', 'OPS', 'ISO', 'BABIP', 'playerID']
bt = bt[bt_col_order]
bt.head()
```

```python
# Filling the NaN values of bt that arise from players with problems like no SB attempts
bt.fillna(0, inplace=True)
bt.head()
```

```python
# hof_ is a dataframe for all HoFamers
hof_ = pd.read_csv("HallOfFame.csv")
hof_ = hof_[hof_["category"] == "Player"].reset_index(drop=True)
hof_["playerID2"] = hof_["playerID"]
hof_.rename(columns = {"playerID" : "name"}, inplace = True)
hof_.rename(columns = {"playerID2" : "playerID"}, inplace = True)
hof_ = hof_[hof_["inducted"] == "Y"].reset_index(drop=True)
hof_ = hof_[hof_["votedBy"] != "Old Timers"].reset_index(drop=True)
hof_["name"].replace(id_dict, inplace=True)
hof_.head()
```

```python
# Making a dictionary and list of all HoFamers
hof_id_list = hof_["playerID"].to_list()
hof_just_names = ppl_[ppl_["playerID"].isin(hof_id_list)].reset_index(drop=True)
hof_just_names = hof_just_names.set_index("playerID")
hof_just_names = hof_just_names.loc[:, ["name"]]

hof_id_dict = hof_just_names.to_dict()
hof_id_dict = hof_id_dict["name"]
hof_id_dict
```

```python
# Creates a dataframe for all HoFamers and their offensive stats
hof_htr_ = bt[bt["playerID"].isin(hof_id_list)].reset_index(drop=True)
hof_htr_.head()
```

```python
# Refining hof to just position player
hof_htr = hof_htr_[hof_htr_["POS"] != "P"].reset_index(drop=True)
hof_htr.head()
```

```python
# hof_btr_stats is a dictionary for the stats of a typical HoFamer
hof_ba = (hof_htr["H"].sum() / hof_htr["AB"].sum()).round(3)
hof_obp = ((hof_htr["H"].sum() + hof_htr["BB"].sum() + hof_htr["HBP"].sum())
           / (hof_htr["AB"].sum() + hof_htr["BB"].sum() + hof_htr["HBP"].sum()
              + hof_htr["SF"].sum())).round(3)
hof_slg = ((hof_htr["1B"].sum() + 2*hof_htr["2B"].sum() + 3*hof_htr["3B"].sum() + 4*hof_htr["HR"].sum())
           / hof_htr["AB"].sum()).round(3)
hof_ops = hof_obp + hof_slg
hof_iso = hof_slg - hof_ba
hof_babip = ((hof_htr["H"].sum() - hof_htr["HR"].sum())
             / (hof_htr["AB"].sum() - hof_htr["K"].sum() - hof_htr["HR"].sum() + hof_htr["SF"].sum())).round(3)
hof_sbp = (hof_htr["SB"].sum() / (hof_htr["SB"].sum() + hof_htr["CS"].sum())).round(3)
hof_btr_stats = {"BA": hof_ba, "OBP": hof_obp, "SLG": hof_slg, "OPS": hof_ops, "ISO": hof_iso,
                 "BABIP": hof_babip, "SB%": hof_sbp}
hof_btr_stats
```

```python
# career_htr has career stats for all MLB position players 1961-
career_htr = bt.groupby(['playerID']).sum()
career_htr.head()
```

```python
# Adding player names to career_htr
career_htr["playerID2"] = career_htr.index
career_htr["name"] = career_htr.index
career_htr = career_htr.replace({"name": id_dict})
career_htr = career_htr.set_index(["name"])
career_htr.rename(columns = {"playerID2" : "playerID"}, inplace = True)
career_htr.head()
```

```python
# career_htr["firstYr"] = bt.groupby(['playerID']).first()["yearID"]
# career_htr["finalYr"] = bt.groupby(['playerID']).last()["yearID"]
# career_htr[["firstYr", "finalYr"]] = career_htr[["firstYr", "finalYr"]].astype(str) 
# career_htr["years"] = career_htr["firstYr"] + " - " + career_htr["finalYr"]
# career_htr

# Adding career span to career_htr
firstYr_dict = bt.groupby(['playerID']).first()["yearID"].to_dict()
finalYr_dict = bt.groupby(['playerID']).last()["yearID"].to_dict()

career_htr["firstYr"] = career_htr["playerID"]
career_htr["finalYr"] = career_htr["playerID"]

career_htr = career_htr.replace({"firstYr": firstYr_dict, "finalYr": finalYr_dict})
career_htr["years"] = career_htr["firstYr"].astype(str) + " - " + career_htr["finalYr"].astype(str)
career_htr.head()
```

```python
# Fixing percentage stats for career_htr
career_htr["BA"] = (career_htr["H"] / career_htr["AB"]).round(3)
career_htr["OBP"] = ((career_htr["H"] + career_htr["BB"] + career_htr["HBP"])
                   / (career_htr["AB"] + career_htr["BB"] + career_htr["HBP"] + career_htr["SF"])).round(3)
career_htr["SLG"] = ((career_htr["1B"] + 2*career_htr["2B"] + 3*career_htr["3B"] + 4*career_htr["HR"])
                     / career_htr["AB"]).round(3)
career_htr["OPS"] = (career_htr["OBP"] + career_htr["SLG"]).round(3)
career_htr["ISO"] = career_htr["SLG"] - career_htr["BA"]
career_htr["BABIP"] = ((career_htr["H"] - career_htr["HR"])
                    / (career_htr["AB"] - career_htr["K"] - career_htr["HR"] + career_htr["SF"])).round(3)
career_htr["SB%"] = (career_htr["SB"] / (career_htr["SB"] + career_htr["CS"])).round(3)
career_htr.fillna(0, inplace=True)
career_htr.head()
```

```python
# Removing the column yearID, which became meaningless
career_htr.drop(["yearID"], axis=1, inplace=True)
career_htr
```

```python
# Adding positions to career_htr
career_htr["POS"] = career_htr["playerID"]
career_htr.replace({"POS": pos_dict}, inplace=True)
career_htr
```

```python
# Finding players whose career stats are on par with HoFamers
hof_htr_comp = career_htr[(career_htr["BA"] >= hof_btr_stats["BA"] - 0.01) &
          (career_htr["OBP"] >= hof_btr_stats["OBP"] - 0.01) &
          (career_htr["SLG"] >= hof_btr_stats["SLG"] - 0.02) &
          (career_htr["OPS"] >= hof_btr_stats["OPS"] - 0.02) &
          (career_htr["ISO"] >= hof_btr_stats["ISO"] - 0.01) &
          (career_htr["BABIP"] >= hof_btr_stats["BABIP"] - 0.02) &
          (career_htr["AB"] >= 7500)]
hof_htr_comp
```

```python
# hof_comp_list is a list of all the players who have HoF-akin stats but are not in the HoF
hof_htr_comp_list = hof_htr_comp["playerID"]
hof_htr_comp_list = list(set(hof_htr_comp_list) - set(hof_id_list))
hof_htr_comp_list
```

```python
# hof_htr_comp is a datframe of players with comparable statistics to that of HoFamers
hof_htr_comp = hof_htr_comp[hof_htr_comp["playerID"].isin(hof_htr_comp_list)]
hof_htr_comp
```

```python
hof_htr_comp.to_excel("HoF_Comparable_Batters.xlsx")
```

```python
# pt is a dataframe for all hitting seasons of MLB players, 1961 marks beginning of Expansion Era
pt = pd.read_csv("Pitching.csv")
pt = pt[pt["yearID"] >= 1961].reset_index(drop=True)
pt.rename(columns = {"SO" : "K"}, inplace = True)
pt.head()
```

```python
pt["playerID2"] = pt["playerID"]
pt.rename(columns = {"playerID": "name", "playerID2" : "playerID"}, inplace = True)
pt = pt.replace({"name": id_dict})
pt["POS"] = ""
pt["POS"] = pt["playerID"]
pt = pt.replace({"POS": pos_dict})
pt.head()
```

```python
# Reordering cols of pt
pt_order = ['name', 'yearID', 'teamID', 'lgID', 'G', 'GS', 'W', 'L', 'CG',
       'SHO', 'SV', 'IPouts', 'H', 'ER', 'HR', 'BB', 'K', 'BAOpp', 'ERA',
       'IBB', 'WP', 'HBP', 'BK', 'BFP', 'GF', 'R', 'SH', 'SF', 'GIDP', 'playerID']
pt = pt[pt_order]
pt.head()
```

```python
# Filling the NaN values of pt columns BAOpp and ERA that arise from players with problems like no outs recorded
pt.fillna({'BAOpp': 0,'ERA': 0}, inplace=True)
pt.head()
```

```python
# Cleaning up pt
pt["IBB"] = pt["IBB"].astype(int)
pt["HBP"] = pt["HBP"].astype(int)
pt["BFP"] = pt["BFP"].astype(int)
pt.head()
```

```python
# Adding stats to pt: H9, HR9, BB9, K9, K/W
pt["WHIP"] = ((pt["BB"] + pt["H"]) / (pt["IPouts"] / 3)).round(3)
pt["H/9"] = (pt["H"] / (pt["IPouts"] / 27)).round(3)
pt["HR/9"] = (pt["HR"] / (pt["IPouts"] / 27)).round(3)
pt["BB%"] = (pt["BB"] / pt["BFP"]).round(3)
pt["K%"] = (pt["K"] / pt["BFP"]).round(3)
pt["K/BB"] = (pt["K"] / pt["BB"]).round(3)
pt.head()
```

```python
# Reordering the cols of bt to a more intuitive order
pt_col_order = ['name', 'yearID', 'teamID', 'lgID', 'G', 'GS', 'W', 'L', 'CG',
                'SHO', 'SV', 'IPouts', 'H', 'ER', 'HR', 'BB', 'K', 'IBB', 'WP',
                'HBP', 'BK', 'BFP', 'GF', 'R', 'SH', 'SF','GIDP', 'BAOpp', 'ERA',
                'WHIP', 'H/9', 'HR/9', 'BB%', 'K%', 'K/BB','playerID']
pt = pt[pt_col_order]
pt.head()
```

```python
# Creates a dataframe for all HoFamers and their pitching stats
hof_ptr = pt[pt["playerID"].isin(hof_id_list)].reset_index(drop=True)
hof_ptr
```

```python
# hof_ptr_stats is a dictionary for the stats of a typical HoFamer
hof_era = ((hof_ptr["ER"].sum() * 9) / (hof_ptr["IPouts"] / 3).sum()).round(2)
hof_whip = ((hof_ptr["BB"].sum() + hof_ptr["H"].sum()) / (hof_ptr["IPouts"].sum() / 3)).round(3)
hof_h9 = (hof_ptr["H"].sum() / (hof_ptr["IPouts"].sum() / 27)).round(3)
hof_hr9 = (hof_ptr["HR"].sum() / (hof_ptr["IPouts"].sum() / 27)).round(3)
hof_bbp = (hof_ptr["BB"].sum() / hof_ptr["BFP"].sum()).round(3)
hof_kp = (hof_ptr["K"].sum() / hof_ptr["BFP"].sum()).round(3)
hof_kbb = (hof_ptr["K"].sum() / hof_ptr["BB"].sum()).round(3)
hof_ptr_stats = {"ERA": hof_era, "WHIP": hof_whip, "H/9": hof_h9, "HR/9": hof_hr9, "BB%": hof_bbp,
                 "K%": hof_kp, "K/BB": hof_kbb}
hof_ptr_stats
```

```python
# career_ptr has career stats for all MLB pitchers 1961-
career_ptr = pt.groupby(['playerID']).sum()
career_ptr.head()
```

```python
# Adding player names to career_ptr
career_ptr["playerID2"] = career_ptr.index
career_ptr["name"] = career_ptr.index
career_ptr = career_ptr.replace({"name": id_dict})
career_ptr = career_ptr.set_index(["name"])
career_ptr.rename(columns = {"playerID2" : "playerID"}, inplace = True)
career_ptr.head()
```

```python
# Adding career span to career_ptr
firstYr_dict = pt.groupby(['playerID']).first()["yearID"].to_dict()
finalYr_dict = pt.groupby(['playerID']).last()["yearID"].to_dict()

career_ptr["firstYr"] = career_ptr["playerID"]
career_ptr["finalYr"] = career_ptr["playerID"]

career_ptr = career_ptr.replace({"firstYr": firstYr_dict, "finalYr": finalYr_dict})
career_ptr["years"] = career_ptr["firstYr"].astype(str) + " - " + career_ptr["finalYr"].astype(str)
career_ptr.head()
```

```python
# Fixing percentage stats for career_ptr
career_ptr["ERA"] = ((career_ptr["ER"] * 9) / (career_ptr["IPouts"] / 3)).round(2)
career_ptr["WHIP"] = ((career_ptr["BB"] + career_ptr["H"]) / (career_ptr["IPouts"] / 3)).round(3)
career_ptr["H/9"] = (career_ptr["H"] / (career_ptr["IPouts"] / 27)).round(3)
career_ptr["HR/9"] = (career_ptr["HR"] / (career_ptr["IPouts"] / 27)).round(3)
career_ptr["BB%"] = (career_ptr["BB"] / career_ptr["BFP"]).round(3)
career_ptr["K%"] = (career_ptr["K"] / career_ptr["BFP"]).round(3)
career_ptr["K/BB"] = (career_ptr["K"] / career_ptr["BB"]).round(3)

career_ptr["ERA"].fillna(0, inplace=True)
career_ptr["WHIP"].fillna(0, inplace=True)
career_ptr["H/9"].fillna(0, inplace=True)
career_ptr["HR/9"].fillna(0, inplace=True)
career_ptr["BB%"].fillna(0, inplace=True)
career_ptr["K%"].fillna(0, inplace=True)
career_ptr["K/BB"].fillna(0, inplace=True)
career_ptr.head()
```

```python
# Removing the columns yearID and BAOpp, which became meaningless
career_ptr.drop(["yearID", "BAOpp"], axis=1, inplace=True)
career_ptr.head()
```

```python
# Cleaning up pt part 2
career_ptr["SH"] = career_ptr["SH"].astype(int)
career_ptr["SF"] = career_ptr["SF"].astype(int)
career_ptr["GIDP"] = career_ptr["GIDP"].astype(int)
career_ptr.head()
```

```python
# Finding players whose career stats are on par with HoFamers
hof_ptr_comp = career_ptr[(career_ptr["ERA"] <= hof_ptr_stats["ERA"] + 0.2) &
                          (career_ptr["WHIP"] <= hof_ptr_stats["WHIP"] + 0.1) &
                          (career_ptr["H/9"] <= hof_ptr_stats["H/9"] + 0.25) &
                          (career_ptr["HR/9"] <= hof_ptr_stats["HR/9"] + 0.1) &
                          (career_ptr["BB%"] <= hof_ptr_stats["BB%"] + 0.01) &
                          (career_ptr["K%"] >= hof_ptr_stats["K%"] - 0.05) &
                          (career_ptr["K/BB"] >= hof_ptr_stats["K/BB"] - 0.5) &
                          (career_ptr["IPouts"] >= 3000)]
hof_ptr_comp
```

```python
# hof_ptr_comp_list is a list of all the players who have HoF-akin stats but are not in the HoF
hof_ptr_comp_list = hof_ptr_comp["playerID"]
hof_ptr_comp_list = list(set(hof_ptr_comp_list) - set(hof_id_list))
hof_ptr_comp_list
```

```python
# hof_ptr_comp is a datframe of players with comparable statistics to that of HoFamers
hof_ptr_comp = hof_ptr_comp[hof_ptr_comp["playerID"].isin(hof_ptr_comp_list)]
hof_ptr_comp
```
