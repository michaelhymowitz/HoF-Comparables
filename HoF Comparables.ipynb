Part 1: Making dictionaries for playerID to player name and playerID to position

```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
# ppl is a dataframe for every MLB players' names
ppl_ = pd.read_csv("People.csv") # ppl_ is created to make later work easier
keep_cols = ["playerID", "nameGiven", "nameLast"]
ppl_ = ppl_.loc[:, keep_cols]
ppl_["name"] = ppl_["nameGiven"] + " " + ppl_["nameLast"]

ppl = ppl_.set_index("playerID")
ppl = ppl.loc[:, ["name"]]
ppl.head()
```

```python
# Making a dictionary and list of all HoFamers
id_dict = ppl.to_dict()
id_dict = id_dict["name"]
id_list = id_dict.keys()
id_dict
```

```python
# fd_dict will let us create separate dataframes for position players and pitchers
# 1961 marks beginning of Expansion Era
pos = pd.read_csv("Fielding.csv")
pos = pos[pos["yearID"] >= 1961].reset_index(drop=True)
pos = pos.loc[:, ["playerID", "POS"]]
pos = pos.set_index("playerID")
pos_dict = pos.to_dict()
pos_dict = pos_dict["POS"]
pos_dict
```

Part 2a: Making a dataframe for all batting seasons post-1961

```python
# bt is a dataframe for all hitting seasons of MLB players post-1961
bt = pd.read_csv("Batting.csv")
bt = bt[bt["yearID"] >= 1961].reset_index(drop=True)
bt.head()
```

```python
bt["playerID2"] = bt["playerID"]
bt.rename(columns = {"playerID": "name", "playerID2" : "playerID"}, inplace = True)
bt.replace({"name": id_dict}, inplace = True)
bt["POS"] = ""
bt["POS"] = bt["playerID"]
bt.replace({"POS": pos_dict}, inplace = True)
bt.head()
```

```python
# Cleaning up bt and adding stats
bt.rename(columns = {"SO" : "K"}, inplace = True)
bt["RBI"] = bt["RBI"].astype(int)
bt["SB"] = bt["SB"].astype(int)
bt["CS"] = bt["CS"].astype(int)
bt["K"] = bt["K"].astype(int)
bt["IBB"] = bt["IBB"].astype(int)
bt["HBP"] = bt["HBP"].astype(int)
bt["SH"] = bt["SH"].astype(int)
bt["SF"] = bt["SF"].astype(int)
bt["GIDP"] = bt["GIDP"].astype(int)
bt.head()
```

```python
# Adding stats to bt
bt["1B"] = bt["H"] - (bt["2B"] + bt["3B"] + bt["HR"])
bt["uBB"] = bt["BB"] - bt["IBB"]
bt["BA"] = (bt["H"] / bt["AB"]).round(3)
bt["OBP"] = ((bt["H"] + bt["BB"] + bt["HBP"])
                   / (bt["AB"] + bt["BB"] + bt["HBP"] + bt["SF"])).round(3)
bt["SLG"] = ((bt["1B"] + 2*bt["2B"] + 3*bt["3B"] + 4*bt["HR"]) / bt["AB"]).round(3)
bt["OPS"] = (bt["OBP"] + bt["SLG"]).round(3)
bt["ISO"] = bt["SLG"] - bt["BA"]
bt["BABIP"] = ((bt["H"] - bt["HR"])
                    / (bt["AB"] - bt["K"] - bt["HR"] + bt["SF"])).round(3)
bt["SB%"] = (bt["SB"] / (bt["SB"] + bt["CS"])).round(3)
bt.head()
```

```python
# Reordering the cols of bt to a more intuitive order
bt_col_order = ['name', 'yearID', 'POS', 'teamID', 'lgID', 'G', 'AB', 'R', 'H', '1B', '2B',
                     '3B', 'HR', 'RBI', 'K', 'BB', 'uBB', 'IBB', 'HBP', 'SH', 'SF', 'GIDP', 'SB',
                     'CS', 'SB%', 'BA', 'OBP', 'SLG', 'OPS', 'ISO', 'BABIP', 'playerID']
bt = bt[bt_col_order]
bt.head()
```

```python
# Filling the NaN values of bt that arise from players with problems like no SB attempts
bt.fillna(0, inplace=True)
bt.head()
```

Part 2b: Making a dataframe and assembling batting statistics for all non-pitcher HoFers

```python
# hof_ is a dataframe for all HoFamers
hof_ = pd.read_csv("HallOfFame.csv")
hof_ = hof_[(hof_["category"] == "Player") & (hof_["inducted"] == "Y")
            & (hof_["votedBy"] != "Old Timers")].reset_index(drop=True)
hof_["playerID2"] = hof_["playerID"]
hof_.rename(columns = {"playerID" : "name"}, inplace = True)
hof_.rename(columns = {"playerID2" : "playerID"}, inplace = True)
hof_["name"].replace(id_dict, inplace=True)
hof_.head()
```

```python
# Making a dictionary and list of all HoFamers
hof_id_list = hof_["playerID"].to_list()
hof_just_names = ppl_[ppl_["playerID"].isin(hof_id_list)].reset_index(drop=True)
hof_just_names = hof_just_names.set_index("playerID")
hof_just_names = hof_just_names.loc[:, ["name"]]

hof_id_dict = hof_just_names.to_dict()
hof_id_dict = hof_id_dict["name"]
hof_id_dict
```

```python
# Creates a dataframe for all HoFamers and their offensive stats
hof_bt_ = bt[bt["playerID"].isin(hof_id_list)].reset_index(drop=True)
hof_bt = hof_bt_[hof_bt_["POS"] != "P"].reset_index(drop=True)
hof_bt_.head()
```

```python
# hof_btr_stats is a dictionary for the stats of a typical HoFamer
hof_ba = (hof_bt["H"].sum() / hof_bt["AB"].sum()).round(3)
hof_obp = ((hof_bt["H"].sum() + hof_bt["BB"].sum() + hof_bt["HBP"].sum())
           / (hof_bt["AB"].sum() + hof_bt["BB"].sum() + hof_bt["HBP"].sum()
              + hof_bt["SF"].sum())).round(3)
hof_slg = ((hof_bt["1B"].sum() + 2*hof_bt["2B"].sum() + 3*hof_bt["3B"].sum() + 4*hof_bt["HR"].sum())
           / hof_bt["AB"].sum()).round(3)
hof_ops = hof_obp + hof_slg
hof_iso = hof_slg - hof_ba
hof_babip = ((hof_bt["H"].sum() - hof_bt["HR"].sum())
             / (hof_bt["AB"].sum() - hof_bt["K"].sum() - hof_bt["HR"].sum() + hof_bt["SF"].sum())).round(3)
hof_sbp = (hof_bt["SB"].sum() / (hof_bt["SB"].sum() + hof_bt["CS"].sum())).round(3)
hof_btr_stats = {"BA": hof_ba, "OBP": hof_obp, "SLG": hof_slg, "OPS": hof_ops, "ISO": hof_iso,
                 "BABIP": hof_babip, "SB%": hof_sbp}
hof_btr_stats
```

Part 2c: Making a dataframe for career batting statistics

```python
# career_bt has career stats for all MLB position players 1961-
career_bt = bt.groupby(['playerID']).sum()
career_bt.head()
```

```python
# Adding player names to career_bt
career_bt["playerID2"] = career_bt.index
career_bt["name"] = career_bt.index
career_bt.replace({"name": id_dict}, inplace = True)
career_bt = career_bt.set_index(["name"])
career_bt.rename(columns = {"playerID2" : "playerID"}, inplace = True)
career_bt.drop(["yearID"], axis=1, inplace=True) # yearID is now meaningless
career_bt.head()
```

```python
# Adding career span to career_bt
firstYr_dict = bt.groupby(['playerID']).first()["yearID"].to_dict()
finalYr_dict = bt.groupby(['playerID']).last()["yearID"].to_dict()

career_bt["firstYr"] = career_bt["playerID"]
career_bt["finalYr"] = career_bt["playerID"]

career_bt.replace({"firstYr": firstYr_dict, "finalYr": finalYr_dict}, inplace = True)
career_bt["years"] = career_bt["firstYr"].astype(str) + " - " + career_bt["finalYr"].astype(str)
career_bt.head()
```

```python
# Fixing percentage stats for career_bt
career_bt["BA"] = (career_bt["H"] / career_bt["AB"]).round(3)
career_bt["OBP"] = ((career_bt["H"] + career_bt["BB"] + career_bt["HBP"])
                   / (career_bt["AB"] + career_bt["BB"] + career_bt["HBP"] + career_bt["SF"])).round(3)
career_bt["SLG"] = ((career_bt["1B"] + 2*career_bt["2B"] + 3*career_bt["3B"] + 4*career_bt["HR"])
                     / career_bt["AB"]).round(3)
career_bt["OPS"] = (career_bt["OBP"] + career_bt["SLG"]).round(3)
career_bt["ISO"] = career_bt["SLG"] - career_bt["BA"]
career_bt["BABIP"] = ((career_bt["H"] - career_bt["HR"])
                    / (career_bt["AB"] - career_bt["K"] - career_bt["HR"] + career_bt["SF"])).round(3)
career_bt["SB%"] = (career_bt["SB"] / (career_bt["SB"] + career_bt["CS"])).round(3)
career_bt.fillna(0, inplace=True)
career_bt.head()
```

```python
# Adding positions to career_bt
career_bt["POS"] = career_bt["playerID"]
career_bt.replace({"POS": pos_dict}, inplace=True)
career_bt
```

Part 2d: Finding HoF comparable batters

```python
# Finding players whose career stats are on par with HoFamers
hof_bt_comp = career_bt[(career_bt["BA"] >= hof_btr_stats["BA"] - 0.01) &
          (career_bt["OBP"] >= hof_btr_stats["OBP"] - 0.01) &
          (career_bt["SLG"] >= hof_btr_stats["SLG"] - 0.02) &
          (career_bt["OPS"] >= hof_btr_stats["OPS"] - 0.02) &
          (career_bt["ISO"] >= hof_btr_stats["ISO"] - 0.01) &
          (career_bt["BABIP"] >= hof_btr_stats["BABIP"] - 0.02) &
          (career_bt["AB"] >= 7500)]
hof_bt_comp
```

```python
# hof_comp_list is a list of all the players who have HoF-akin stats but are not in the HoF
hof_bt_comp_list = hof_bt_comp["playerID"]
hof_bt_comp_list = list(set(hof_bt_comp_list) - set(hof_id_list))
hof_bt_comp_list
```

```python
# hof_bt_comp is a datframe of players with comparable statistics to that of HoFamers
hof_bt_comp = hof_bt_comp[hof_bt_comp["playerID"].isin(hof_bt_comp_list)]
hof_bt_comp
```

```python
hof_bt_comp.to_excel("HoF_Comparable_Batters.xlsx")
```

Part 3a: Making a dataframe for all pitching seasons post-1961

```python
# pt is a dataframe for all pitching seasons of MLB players post-1961
pt = pd.read_csv("Pitching.csv")
pt = pt[pt["yearID"] >= 1961].reset_index(drop=True)
pt.head()
```

```python
pt["playerID2"] = pt["playerID"]
pt.rename(columns = {"playerID": "name", "playerID2" : "playerID"}, inplace = True)
pt.replace({"name": id_dict}, inplace = True)
pt["POS"] = ""
pt["POS"] = pt["playerID"]
pt.replace({"POS": pos_dict}, inplace = True)
pt.head()
```

```python
# Filling the NaN values of pt columns BAOpp and ERA that arise from players with problems like no outs recorded
pt.fillna({'BAOpp': 0,'ERA': 0}, inplace=True)
pt.head()
```

```python
# Cleaning up pt
pt.rename(columns = {"SO" : "K"}, inplace = True)
pt["IBB"] = pt["IBB"].astype(int)
pt["HBP"] = pt["HBP"].astype(int)
pt["BFP"] = pt["BFP"].astype(int)
pt.head()
```

```python
# Adding stats to pt: H9, HR9, K9, BB9, K/BB
pt["WHIP"] = ((pt["BB"] + pt["H"]) / (pt["IPouts"] / 3)).round(3)
pt["H/9"] = (pt["H"] / (pt["IPouts"] / 27)).round(3)
pt["HR/9"] = (pt["HR"] / (pt["IPouts"] / 27)).round(3)
pt["K%"] = (pt["K"] / pt["BFP"]).round(3)
pt["BB%"] = (pt["BB"] / pt["BFP"]).round(3)
pt["K/BB"] = (pt["K"] / pt["BB"]).round(3)
pt.head()
```

```python
# Reordering the cols of bt to a more intuitive order
pt_col_order = ['name', 'yearID', 'teamID', 'lgID', 'G', 'GS', 'W', 'L', 'CG',
                'SHO', 'SV', 'IPouts', 'H', 'ER', 'HR', 'BB', 'K', 'IBB', 'WP',
                'HBP', 'BK', 'BFP', 'GF', 'R', 'SH', 'SF','GIDP', 'BAOpp', 'ERA',
                'WHIP', 'H/9', 'HR/9', 'K%', 'BB%', 'K/BB','playerID']
pt = pt[pt_col_order]
pt.head()
```

Part 3b: Making a dataframe and assembling pitching statistics for all pitcher HoFers

```python
# Creates a dataframe for all HoFamers and their pitching stats
hof_pt = pt[pt["playerID"].isin(hof_id_list)].reset_index(drop=True)
hof_pt
```

```python
# hof_pt_stats is a dictionary for the stats of a typical HoFamer
hof_era = ((hof_pt["ER"].sum() * 9) / (hof_pt["IPouts"] / 3).sum()).round(2)
hof_whip = ((hof_pt["BB"].sum() + hof_pt["H"].sum()) / (hof_pt["IPouts"].sum() / 3)).round(3)
hof_h9 = (hof_pt["H"].sum() / (hof_pt["IPouts"].sum() / 27)).round(3)
hof_hr9 = (hof_pt["HR"].sum() / (hof_pt["IPouts"].sum() / 27)).round(3)
hof_kp = (hof_pt["K"].sum() / hof_pt["BFP"].sum()).round(3)
hof_bbp = (hof_pt["BB"].sum() / hof_pt["BFP"].sum()).round(3)
hof_kbb = (hof_pt["K"].sum() / hof_pt["BB"].sum()).round(3)
hof_pt_stats = {"ERA": hof_era, "WHIP": hof_whip, "H/9": hof_h9, "HR/9": hof_hr9, "K%": hof_kp,
                 "BB%": hof_bbp, "K/BB": hof_kbb}
hof_pt_stats
```

Part 3c: Making a dataframe for career pitching statistics

```python
# career_pt has career stats for all MLB pitchers 1961-
career_pt = pt.groupby(['playerID']).sum()
career_pt.head()
```

```python
# Adding player names to career_pt
career_pt["playerID2"] = career_pt.index
career_pt["name"] = career_pt.index
career_pt.replace({"name": id_dict}, inplace = True)
career_pt = career_pt.set_index(["name"])
career_pt.rename(columns = {"playerID2" : "playerID"}, inplace = True)
career_pt.head()
```

```python
# Adding career span to career_pt
firstYr_dict = pt.groupby(['playerID']).first()["yearID"].to_dict()
finalYr_dict = pt.groupby(['playerID']).last()["yearID"].to_dict()

career_pt["firstYr"] = career_pt["playerID"]
career_pt["finalYr"] = career_pt["playerID"]

career_pt.replace({"firstYr": firstYr_dict, "finalYr": finalYr_dict}, inplace = True)
career_pt["years"] = career_pt["firstYr"].astype(str) + " - " + career_pt["finalYr"].astype(str)
career_pt.head()
```

```python
# Removing the columns yearID and BAOpp, which are now meaningless
career_pt.drop(["yearID", "BAOpp"], axis=1, inplace=True)
career_pt.head()
```

```python
# Fixing percentage stats for career_pt
career_pt["ERA"] = ((career_pt["ER"] * 9) / (career_pt["IPouts"] / 3)).round(2)
career_pt["WHIP"] = ((career_pt["BB"] + career_pt["H"]) / (career_pt["IPouts"] / 3)).round(3)
career_pt["H/9"] = (career_pt["H"] / (career_pt["IPouts"] / 27)).round(3)
career_pt["HR/9"] = (career_pt["HR"] / (career_pt["IPouts"] / 27)).round(3)
career_pt["K%"] = (career_pt["K"] / career_pt["BFP"]).round(3)
career_pt["BB%"] = (career_pt["BB"] / career_pt["BFP"]).round(3)
career_pt["K/BB"] = (career_pt["K"] / career_pt["BB"]).round(3)

career_pt["ERA"].fillna(0, inplace=True)
career_pt["WHIP"].fillna(0, inplace=True)
career_pt["H/9"].fillna(0, inplace=True)
career_pt["HR/9"].fillna(0, inplace=True)
career_pt["K%"].fillna(0, inplace=True)
career_pt["BB%"].fillna(0, inplace=True)
career_pt["K/BB"].fillna(0, inplace=True)
career_pt.head()
```

```python
# Cleaning up pt more
career_pt["SH"] = career_pt["SH"].astype(int)
career_pt["SF"] = career_pt["SF"].astype(int)
career_pt["GIDP"] = career_pt["GIDP"].astype(int)
career_pt.head()
```

Part 2d: Finding HoF comparable batters

```python
# Finding players whose career stats are on par with HoFamers
hof_pt_comp = career_pt[(career_pt["ERA"] <= hof_pt_stats["ERA"] + 0.2) &
                          (career_pt["WHIP"] <= hof_pt_stats["WHIP"] + 0.1) &
                          (career_pt["H/9"] <= hof_pt_stats["H/9"] + 0.25) &
                          (career_pt["HR/9"] <= hof_pt_stats["HR/9"] + 0.1) &
                          (career_pt["K%"] >= hof_pt_stats["K%"] - 0.05) &
                          (career_pt["BB%"] <= hof_pt_stats["BB%"] + 0.01) &
                          (career_pt["K/BB"] >= hof_pt_stats["K/BB"] - 0.5) &
                          (career_pt["IPouts"] >= 3000)]
hof_pt_comp
```

```python
# hof_pt_comp_list is a list of all the players who have HoF-akin stats but are not in the HoF
hof_pt_comp_list = hof_pt_comp["playerID"]
hof_pt_comp_list = list(set(hof_pt_comp_list) - set(hof_id_list))
hof_pt_comp_list
```

```python
# hof_pt_comp is a datframe of players with comparable statistics to that of HoFamers
hof_pt_comp = hof_pt_comp[hof_pt_comp["playerID"].isin(hof_pt_comp_list)]
hof_pt_comp
```

```python
hof_pt_comp.to_excel("HoF_Comparable_Pitchers.xlsx")
```
